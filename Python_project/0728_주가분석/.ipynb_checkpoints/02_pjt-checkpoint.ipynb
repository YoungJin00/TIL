```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("archive/NFLX.csv")
df
```

# A. 데이터 전처리 - 데이터 읽어오기 

```python
# iloc 활용하여 슬라이싱을 통해 행은 전체 열을 5번쨰 까지만 불러오기
df = df.iloc[:, :5]
df
```

# B. 데이터 전처리 - 2021년 이후의 종가 데이터 출력하기

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# 2021년 이후의 종가 데이터 출력
df_after_2021 = df[df["Date"] >= "2021-01-01"]
df_after_2021

# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_after_2021['Date'], df_after_2021['Close'])


# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Close')

# 그래프 표시
plt.show()
```

# C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기

```python
# C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기
max_price = max(df_after_2021['Close'])
min_price = min(df_after_2021['Close'])

print(f'최고 종가 : {max_price}')
print(f'최저 종가 : {min_price}')
```

# D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기


```python
# D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기

# 2021년 데이터만 추출하기
df_2021 = df_after_2021[(df_after_2021["Date"] >= "2021-01-01") & (df_after_2021["Date"] < "2022-01-01")]
# 월 컬럼을 만들어 주기
df_2021['month'] = df_2021['Date'].dt.month
# 월별 평균 값 구하기
df_mean = df_2021.groupby(by='month').mean()
df_mean
```

```python
# 2022년 데이터만 추출
df_2022 = df_after_2021[df_after_2021["Date"] >= "2022-01-01"]
df_2022
# 2022년 월 데이터 추출
df_2022['month'] = df_2022['Date'].dt.month
# 2022년 월별 데이터 평균 구하기
df2_mean = df_2022.groupby(by='month').mean()
df2_mean
```

```python
# 2021년 데이터와 2022년 데이터 합치기
df_total = pd.concat([df_mean, df2_mean])
df_total
```

```python
# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_total['Date'], df_total['Close'])

# 그래프 제목 설정
plt.title('Monthly Average Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Average Close price')

# 그래프 표시
plt.show()
```

# E. 데이터 시각화 - 2022년 이후 최고, 최저 , 종가 시각화 하기

```python
# # 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_2022['Date'], df_2022['Low'], label='Low')
plt.plot(df_2022['Date'], df_2022['High'], label='High')
plt.plot(df_2022['Date'], df_2022['Close'], label='Close')

# 그래프 제목 설정
plt.title('High, Low, CLose Prices since January 2022')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Price')

# x 축 설정(회전시키기)
# plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()

```
